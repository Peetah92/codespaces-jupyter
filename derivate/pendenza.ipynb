# Pendenza di una curva

Importazione librerie

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definizione della funzione

```python
def f(x):
    return(x**3-2*x**2-x-3)
```

Verifica

```python
print(f(0.5))
```

Scelgo un valore dx per $\Delta x$

```python
dx=0.00001
```

Definisco un vettore che contenga i valori di $x$ da -10 a 10 ogni 0.01, sono 2001 valori

```python
x=np.linspace(-10,10,2001)
print(x)
```

Calcolo il vettore y delle immagini

```python
y=f(x)
```

# calcolo la pendenza di $f(x)$ per i valori 

```python
m=(f(x+dx)-f(x))/dx
print(m)
```

Rappresento graficamente la funzione $f(x)$ e la sua pendenza $m(x)$

```python
griglia=np.linspace(-10,10,21)

plt.figure(figsize=(6,6))
plt.plot(x,y)                                   #plotta f(x)
plt.plot(x,m)                                   #plotta m(x)
plt.axis([-10,10,-10,10])                       #range assi
plt.gca().set_aspect("equal")                   #scale x e y uguali
plt.xticks(griglia)                             #tacchette griglia orizzontali
plt.yticks(griglia)                             #tacchette griglia verticali
plt.axhline(0, color='black', linewidth=1)      #linea in corrispondenza asse orizzontale
plt.axvline(0, color='black', linewidth=1)      #linea in corrispondenza asse verticale
plt.grid()                                      #griglia
plt.show()                                      #mostra il diagramma
```

eseguiamo una regressione polinomiale su $m(x)$

```python
np.polyfit(x,m,2)
```
