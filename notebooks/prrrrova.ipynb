Importazione delle librerie necessarie

Esempi con la libreria math

```python
import math
from math import cos
from math import acos as arccos
from math import log as ln

import numpy as np

import matplotlib.pyplot as plt
```

```python
m=6.9 # massa di 6.9 kg
print("massa =", m, "kg")
print(math.sin(m))
print(cos(math.pi))
arccos(cos(math.pi))
print("ln(10) = ", ln(10))
```

```python
a=np.array([1,2,3])
print(a)
print(a+1)
print(np.sin(a))
b=np.array([[1,2,3],[4,5,6]])
print(b)
print(np.sin(b))
z=np.zeros(11)
print(z)
z[3]=5
print(z)
z[0]=43
print(z)
bz=np.zeros_like(b)
print(bz)
```

```python
x=np.linspace(1,10,10)
print(x)
x=np.linspace(0,10,11)
print("x = ", x)
y=3*x+2
print("y = ", y)
```
